```python
#reverse a string without in-built functions
def reverse_string(s):
    reversed_s = ""
    for char in s:
        reversed_s = char+reversed_s
    print(s)
    return reversed_s


import time
while True:
    input_str = input("Give a string to reverse:: ")

    if (input_str.lower() != 'quit'):
        # Get and print the response
        response = reverse_string(input_str)
        print("\nreversed string:", response)
        
        # add delay of 3 seconds
        time.sleep(3)
    else:
        break  
```

```python
#program to check if a string is a  palindrome  
def palindrome_chk(s):
    reversed_s = ""
    for char in s:
        reversed_s = char+reversed_s
    if reversed_s == s:
        print("Palindrome")
    else:
        print("Not a Palindrome")

import time
while True:
    input_str = input("Give a string to check if palindrome:: ")

    if (input_str.lower() != 'quit'):
        # Get and print the response
        response = palindrome_chk(input_str)
        # add delay of 3 seconds
        time.sleep(3)
    else:
        break  
```

```python
#Find missing number
def missing_number(arr):
    n = len(arr)+ 1;
    totSum = sum(arr)
    expSum = n * (n + 1) // 2;
    
    #Return the missing number
    print(expSum - totSum)
    return (expSum - totSum);

import time

while True:
    input_string = input("Enter numbers separated by spaces or quit to exit:")
    if (input_string != 'quit'): 
        string_list = input_string.split()
        input_arr = [int(num) for num in string_list] # Using list comprehension for conversion
        response = missing_number(input_arr)
        print("missing number in array:",response)

    else:
        break
        
```

```python
#program to check if a string is a  palindrome  
def cnt_word_frqncy(s):
    #created empty dictionary
    M = {}
    # string for storing the words
    word = ""
    
    for i in range(len(s)):    
        # Check if current character is blank space then it means we have got one word
        if (s[i] == ' '):   
            # If the current word is not found then insert current word with frequency 1
            if (word not in M):
                M[word] = 1
                word = ""
            
            # update the frequency
            else:
                M[word] += 1
                word = ""
        
        else:
            word += s[i]
    
    # Storing the last word of the string
    if (word not in M):
        M[word] = 1
    
    # Update the frequency
    else:
        M[word] += 1

    # Traverse the map to print the frequency
    for it in M:
        print(it, "-", M[it])

import time
while True:
    input_str = input("Give a string to check word frequency: ")

    if (input_str.lower() != 'quit'):
        # Get and print the response
        cnt_word_frqncy(input_str)
        # add delay of 3 seconds
        time.sleep(3)
    else:
        break 
```

```python
#print based on multiples
def multiples_n_print(num):
    if ((num%3) == 0):
        print(num,"Fizz")
    
    if ((num%5) == 0):
        print(num,"Buzz")
    
    if ((num%3) == 0 and (num%5) == 0):
        print(num,"FizzBuzz")

while True:
    input_a = input("Enter a start limit:") 
    input_b = input("Enter end limit:")
    if (input_a != 'quit' or input_b != 'quit'):  
        for i in range(int(input_a), int(input_b)):
            multiples_n_print(i)
    else: 
        break 
```
