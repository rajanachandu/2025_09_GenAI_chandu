```python
#dict
person1 = {
    "name": "John",
    "age": 30,
    "city": "New York"
}

person2 = {
    "name": "Ana",
    "age": 49,
    "city": "London"
}

for person in [person1, person2]:
    print(person)
    print(person["name"])

try:
    del person1["city2"]
except Exception as e:
    print(f"Error in removing key: {e}")

#list
my_list = [1,2,3,4,5]

for num in my_list:
    print(num)
```

```python
#print squares of numbers from list
def square_numbers(l):
    for num in l:
        print(num*num)

my_list = [1,2,3,4,5]
square_numbers(my_list)
```
