```python
#Create a list of squares of all even numbers from 1 to 20.

result = []

def even_squares(start, end):
    """Return a list of squares of all even numbers in the given range."""
    for x in range(start, end + 1):
        if x % 2 == 0:
            result.append(x**2)
            

# Call the function for numbers 1 to 20
even_squares(1, 20)
print(result)
```

```python
#Write a Python function `get_top_k_docs(query, embeddings, k)` that returns top-k cosine-similar documents.
import numpy as np

from sentence_transformers import SentenceTransformer
model = SentenceTransformer("all-MiniLM-L6-v2")

def get_top_k_docs(query, embeddings, k):
    query_vector = model.encode([query])[0]
    similarities = model.similarity(query_vector, embeddings)

    # Ensure it's a 1D numpy array
    similarities = np.asarray(similarities).squeeze()


    # Now get top 3
    top_3_indices = np.argsort(similarities)[::-1][:k]
    top_scores = similarities[top_3_indices]
    return top_scores
```

```python
#Write a function that reads a text file and returns its length, handling the case where the file doesnâ€™t exist.
def get_file_length(file_path):
    """
    Reads a text file and returns its length (number of characters).
    If the file does not exist, returns None and prints an error message.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            return len(content)
    except FileNotFoundError:
            print(f"Error: The file '{file_path}' does not exist.")
            return None

# Example usage
file_path = "/Users/chandurajana/Documents/Chandu_Training/2025_09_GenAI_chandu/session4/read_txt_file.txt"
length = get_file_length(file_path)

if length is not None:
    print(f"The length of the file is {length} characters.")
```

```python
#Generate random 5D vectors using NumPy and compute pairwise cosine similarity between them.
import numpy as np

# Step 1: Random 4x5 vectors
vectors = np.random.rand(4,5)
print("4 Random 5D vector:", vectors)

# Step 2: Normalize
norms = np.linalg.norm(vectors, axis=1, keepdims=True)
vectors_normalized = vectors / norms

# Step 3: Cosine similarity
cosine_similarity = np.dot(vectors_normalized, vectors_normalized.T)
print("\nPairwise Cosine Similarity:\n", cosine_similarity)
```
