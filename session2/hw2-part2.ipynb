```python
#3. Chat Completion
#Write Python code to:
#Take a user question as input.
#Send it to the Chat Completion API (gpt-4.1-mini or gpt-5-mini).
#Print the AI’s answer.

import os
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
my_api_key=os.getenv("OPEN_AI_API_KEY")
print(my_api_key)
```

```python
#pip install openai
from openai import OpenAI
client = OpenAI(api_key=my_api_key)

prompt_responses = []

def ask_question(prompt):
    print(f"User asked: {prompt}")
    response = client.chat.completions.create(
        model="gpt-5-mini",
        messages=[
            {"role":"system", "content": "You are a helpful assistant. Answer as concisely as possible."},
            {"role":"user", "content":prompt}
        ]
    )
    
    prompt_responses.append(response.choices[0].message.content)

    #return response.choices[0].message.content
```

```python
import time

count=0
while count<3:
    user_input = input("Ask something:")
    if (user_input != 'quit'):
        ask_question(user_input)
        count+=1
        #print("\nopenAI says:",response.json()) 
        time.sleep(3)
    else:
        break

for resp in prompt_responses:
    print("\nopenAI says:",resp)
```
