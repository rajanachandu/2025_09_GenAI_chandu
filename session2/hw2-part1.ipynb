```python
#1. Lists
#Create a list of your top 5 favorite movies.
#Print the first and last movie using list indexing.
#Add one more movie to the list and remove one movie.
#Write a loop that prints all movies in uppercase.

#Create a list of your top 5 favorite movies.
movies = ['Alien','Brazil','Ran','Dance','H2o']

#Print the first and last movie using list indexing.
print(movies[0])
print(movies[-1])

#Add one more movie to the list and remove one movie.
movies.append('Baby')
print(movies)
movies.remove('Baby')
print(movies)

#Write a loop that prints all movies in uppercase.
for movie in movies:
    print(movie.upper())
```
