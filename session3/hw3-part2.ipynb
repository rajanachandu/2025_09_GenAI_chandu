```python
# Part 2: Working with Local LLM – Ollama + Mistral
# Problem 1 – Basic Chat App
import subprocess

def ask_model(prompt, model="mistral:instruct"):
    result = subprocess.run(
        ["ollama", "run", model],
        input=prompt.encode("utf-8"),
        capture_output=True
    )
    return result.stdout.decode("utf-8").strip()

response = ask_model("Write a short haiku about AI.")
print(response)
```
