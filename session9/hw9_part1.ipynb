```python
#Given a list of 10 numbers, print sum, max, min, and average.
from statistics import mean, median, mode

def calc(num_list):
        sum_v=sum(num_list)
        avg=mean(num_list)
        med=median(num_list)
        mod=mode(num_list)
        return sum_v, avg, med, mod

num_list = [2,4,7,9,3,1,5,7,8,2]
sum_val, avg_val, med_val, mod_val = calc(num_list)
print("Sum:", sum_val, "Mean:", avg_val, "Median:", med_val, "Mode:", mod_val)
```

```python
#4.Write is_prime(n) that returns True if n is prime.

def chk_prime(num):
    for i in range(2,num):
            if (num%i==0):
                print(f"i value = {i}")
                return False
                
    
    if (i+1==num):
        print(f"i value = {i}")
        return True

num = 20
bool_value = chk_prime(num)
if bool_value:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
#5.Write convert_temp(value, scale) for C↔F conversion.

from pandas.core.internals.blocks import F


def convert_temp(value, scale):
    if (scale=='F'):
        C=5/9*(value-32)
        print(f"We got temperature in Fahrenheit:{value}, In terms of Celsius:{C}")
    elif (scale=='C'):
        F=(9/5*value)+32
        print(f"We got temperature in Celsius:{value}, In terms of Fahrenheit:{F}")

convert_temp(84, 'F')
convert_temp(23, 'C')
```

```python
#6.Use list comprehension to create squares of all odd numbers 1–20.

def sqr_odd_num(num):
    if (num%2!=0):
        print(f"{num*num}")

for i in range(1,21):
    sqr_odd_num(i)

#Method 2
squares = [n*n for n in range(1,21) if n%2!=0]
print(squares)
```

```python
#7.Create class Task(name, priority, status) with method mark_complete(). Instantiate 3 tasks, mark one done.

class Task:
    def __init__(self, name, priority, status):
        self.name = name
        self.priority = priority
        self.status = status

    def mark_complete(self):
        
        print(f"Task:{self.name}, Priority:{self.priority}, status:{self.status}")

cooking = Task("cooking",1,"half done")
cooking.mark_complete()
laundry = Task("laundry",2,"done")
laundry.mark_complete()
dishes =  Task("dishes",3,"not done")
dishes.mark_complete()


```
